## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
N = int(input("Enter a number: "))
even_count = 0
odd_count = 0
for i in range(1, N + 1):
    if i % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
print("Even numbers count:", even_count)
print("Odd numbers count:", odd_count)

```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
num = int(input("Enter a number: "))
is_prime = True
if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
N = int(input("Enter a number: "))

print("Prime numbers between 1 and", N, "are:")

for num in range(2, N + 1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break


    if is_prime:
        print(num, end=" ")

```

## Question 18 (Hard)

**Title:** First Number Divisible by 7 and 11

### Problem Statement:
Take N as input and find the first number between 1 and N that is divisible by both 7 and 11. Stop after finding it.

### Approach:
Use logical AND and stop the loop using `break`.


```python
N = int(input("Enter a number: "))

for i in range(1, N + 1):

    if i % 7 == 0 and i % 11 == 0:
        print("The first number divisible by both 7 and 11 is:", i)
        break
else:
    print("No number between 1 and", N, "is divisible by both 7 and 11.")

```
