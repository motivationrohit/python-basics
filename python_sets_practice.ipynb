### Q5
Given:
```python
A = {2, 4, 6, 8}
B = {4, 8, 12}
```
Find Union, Intersection, A-B

````python
A = {2, 4, 6, 8}
B = {4, 8, 12}
x=set(A)
y=set(B)
print(x.union(y))
print(x.intersection(y))
print(x.union(y))
````
