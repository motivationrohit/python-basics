Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
i = 1
while i <= 20:
    print(i)
    i += 1

```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
n = int(input("Enter a number: "))
total = 0
i = 1
while i <= n:
    total += i
    i += 1

print("Sum =", total)
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
var= int(input())

new = var

res=0

while var>0:
  temp = var % 10
  res = res*10 + temp
  var = var // 10

print(res)


if new==res:
  print("Palindrome")
else:
  print("Not Palindrome")


```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
num = int(input("Enter a number: "))
count = 0
if num == 0:
    count = 1
else:

    while num > 0:
        num //= 10
        count += 1
print("Number of digits:", count)
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
num = int(input("Enter a number: "))
original = num
reverse_num = 0
while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num //= 10


if original == reverse_num:
    print("Palindrome number")
else:
    print("Not a palindrome number")


```
