### Q33. Flatten [[1,2,3],[4,5,6],[7,8,9]].

```python
a=[[1,2,3],[4,5,6],[7,8,9]]
flat=[x for item in a for x in (item if isinstance(item,list) else [item])]
print(flat)
```

```python
a=[[1,2,3],[4,5,6],[7,8,9]]
flat=[]
for i in a:
  flat.append(i)
print(flat)
```

```python
a=[[1,2,3],[4,5,6],[7,8,9]]
flat=[]
for i in a:
  if isinstance(i,list):
    flat.extend(i)
  else:
    flat.append(i)
print(flat)
```
